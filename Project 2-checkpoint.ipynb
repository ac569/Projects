```python
from imblearn.over_sampling import SMOTE
```

```python
from collections import Counter
```

```python
y=LabelEncoder().fit_transform(y)
```

```python
oversample=SMOTE()
```

```python
from matplotlib import pyplot
x, y = oversample.fit_resample(x, y)
# summarize distribution
counter = Counter(y)
for k,v in counter.items():
	per = v / len(y) * 100
	print('Class=%d, n=%d (%.3f%%)' % (k, v, per))
# plot the distribution
plt.bar(counter.keys(), counter.values())
plt.show()
```

```python
#we changed the data distribution ny SMOTE oversampling and made all six classes in equal distribution
```

* We could have under sampled the data as well however we will end up loosing data in the process

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=True)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
# we will directly standardise the data before we go for knn model
from sklearn.preprocessing import StandardScaler
st=StandardScaler()
x_train_scaled=st.fit_transform(x_train)
x_test_scaled=st.fit_transform(x_test)
```

# KNN Model

```python
#we will check for which values of k the error in the model is less as it will help us getting better knn model
```

```python
error_rate=[]
for i in range(1,10):
             knn = KNeighborsClassifier(n_neighbors=i)
             model = knn.fit(x_train_scaled,y_train)
             pred_i = knn.predict(x_test_scaled)
             error_rate.append(np.mean(pred_i != y_test))
plt.figure(figsize=(13,8))
plt.plot(range(1,10), error_rate, linestyle = 'dotted', marker = 'o',color = 'b')
plt.xlabel('K value')
plt.ylabel('Error Rate')
plt.title('K value Vs Error Rate')
plt.show()
```

* As we can see lower the k value lower is the error rate
* However to avoid overfitting we will avoid using k=1,2 and go with k=3,4

```python
model = KNeighborsClassifier(n_neighbors=3).fit(x_train_scaled,y_train)
pred = model.predict(x_test_scaled)
plot_confusion_matrix(model,x_test_scaled,y_test,cmap=plt.cm.Blues)
print(classification_report(y_test,pred))
print('The score of training knn model for k=3:',model.score(x_train_scaled,y_train))
print('The score of testing knn model for k=3 :', model.score(x_test_scaled,y_test))
```

```python
model = KNeighborsClassifier(n_neighbors=4).fit(x_train_scaled,y_train)
pred = model.predict(x_test_scaled)
plot_confusion_matrix(model,x_test_scaled,y_test,cmap=plt.cm.Blues)
print(classification_report(y_test,pred))
print('The score of training knn model for k=4:',model.score(x_train_scaled,y_train))
print('The score of testing knn model for k=4 :', model.score(x_test_scaled,y_test))
```

* As it can be seen from above the error rate is lesser when k=3 than when k=4.
* The precision,recall,f1 score are all better for k=3.
* The score of the model also declines with increase in values of k. 

# Conclusion
* From the above results it can be said that k=3 will give us the best result for the parameters we have chosen.
* Except for class 3 and 2 all the other classes have better accuracies.
* This could be because of oversampling as it replicates data and overfitting of data is possible, but the same could be said for undersampling which could have lead to data loss.
* We can improve the model by taking other parameters like weights, other normalizing or scaling techniques or changing the random state.
* We can use other models like logistic regression ,Decision Tree to improve the model and its accuracy.
* With increase in k value we do see that the gap between training and testing scores decrease which is a good sign and reduces overfitting.
